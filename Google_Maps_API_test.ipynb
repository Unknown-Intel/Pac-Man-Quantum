```python
# import pycountry as countries
from countryinfo import CountryInfo
from PIL import Image
import numpy as np
import requests
import shutil
countries = CountryInfo()

x = 0
y = 0
width = 100
height = 90

basewidth = 28
hsize = 26

api_key = "AIzaSyC7CEkAND6bNxkwUi4POEHg4XAZ90TsTpU"
size = "100x100"
zoom = 14
scale = 2
feature = "all"
element = "all"
visibility = "off"
color = "0xffffff"
feature_road = "road"
element_road = "geometry"
visibility_road = "simplified"
color_road = "0x000000"
weight_road = "2"
static_url = "https://maps.googleapis.com/maps/api/staticmap?"
style = "feature:" + feature + "|element:" + element + "|visibility:" + visibility + "|color:" + color
style_road = "feature:" + feature_road + "|element:" + element_road + "|visibility:" + visibility_road + "|color:" + color_road + "|weight:" + weight_road

for country in countries.all():
    try :
        name = CountryInfo(country).capital()
    except:
        continue
        
    r = requests.get(static_url + "center=" + str(name) + "&zoom=" + str(zoom) + "&size=" + size + "&scale=" + str(scale) + "&style=" + style + "&style=" + style_road +"&key=" + api_key)
    if r.status_code == 200:
        with open("./resource/generated_images/img_" + str(name) + ".png", 'wb') as f:
            f.write(r.content)
            
    image = Image.open('./resource/generated_images/img_' + str(name) + '.png')
    temp = image.crop((x,y, x + width ,y + height))
    temp = temp.resize((basewidth,hsize), Image.LANCZOS)
    temp.save('./resource/generated_images/img_' + str(name) + '.png')
#     temp.show()
    arr = np.array(temp)
    counts = 0
    text_image = ""
    for i in range(len(arr)):
        temp = ""
        for j in range(len(arr[i])):
            if arr[i][j] >= 50 :
                temp += "_"
                counts += 1
            else:
                temp += "#"
        temp += '\n'
        text_image += temp
    if counts / (basewidth * hsize) >= 0.25 and counts / (basewidth * hsize) <= 0.50:   
        newLoc = "./resource/generated_maps/test" + str(name) + ".txt"
        text_file = open(newLoc , "w")
        text_file.write(text_image)
        text_file.close()
        print("Completed : ", name)
```
